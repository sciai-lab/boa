```python
# for each experiment in experiment_dict, load the 
from pathlib import Path

import numpy as np
import torch


for experiment, params in experiment_dict.items():
    print(f"Experiment: {experiment}")
    experiment_dict[experiment]["nmapes"] = []
    for test_path in params["test_paths"]:
        nmape_path = Path(test_path) / "tensorboard/version_0/nmape_test.pt"
        if nmape_path.exists():
            nmape = np.array(torch.load(open(nmape_path, "rb")))
            experiment_dict[experiment]["nmapes"].append(nmape)
            print(f"  Test path: {test_path}, NMAPE: {np.mean(nmape):.4f} ± {np.std(nmape):.4f}")
```

```python
# for each experiment calculate the mean and std of the means of the nmapes
for experiment, params in experiment_dict.items():
    nmapes = params["nmapes"]
    mean_nmapes = [np.mean(nmape) for nmape in nmapes]
    # standard error of the mean
    sem_nmapes = np.std(mean_nmapes) / np.sqrt(len(mean_nmapes))
    print(f"Experiment: {experiment}, Mean NMAPE: {np.mean(mean_nmapes):.4f} ± {sem_nmapes:.4}")
```

```python
# Plot a histogram of the nmapes for each experiment
import matplotlib.pyplot as plt
for experiment, params in experiment_dict.items():
    nmapes = params["nmapes"]
    plt.figure(figsize=(8, 6))
    for i, nmape in enumerate(nmapes):
        plt.hist(nmape, bins=30, alpha=0.5, label=f"Test {i+1}")
    plt.xlabel("NMAPE")
    plt.ylabel("Frequency")
    plt.title(f"NMAPE Distribution for {experiment}")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

for experiment, params in experiment_dict.items():
    nmapes = params["nmapes"]
    all_nmape = np.concatenate(nmapes)
    # Remove extreme outliers for visualization (e.g., clip to 99th percentile)
    lower, upper = np.percentile(all_nmape, [1, 99])
    clipped_nmape = all_nmape[(all_nmape >= lower) & (all_nmape <= upper)]

    plt.figure(figsize=(8, 6))
    plt.hist(clipped_nmape, bins=30, alpha=0.7, color='tab:blue')
    plt.xlabel("NMAPE")
    plt.ylabel("Frequency")
    plt.title(f"NMAPE Distribution for {experiment} (1st-99th percentile)")
    plt.grid()
    plt.show()

    # Optionally, also show a boxplot for a summary
    plt.figure(figsize=(6, 2))
    plt.boxplot(clipped_nmape, vert=False, showfliers=False)
    plt.xlabel("NMAPE")
    plt.title(f"NMAPE Boxplot for {experiment} (no outliers)")
    plt.grid(axis='x')
    plt.show()
```
